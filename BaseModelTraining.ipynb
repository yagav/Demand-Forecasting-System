```python
import pandas as pd

df = pd.read_csv("data/sales_2013.csv")
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values("date")
```

```python
def create_features(df):
    df = df.sort_values("date")
    df['lag_1'] = df['sales'].shift(1)
    df['lag_7'] = df['sales'].shift(7)
    df['rolling_7'] = df['sales'].rolling(7).mean()
    return df.dropna()
```

```python
import mlflow
import numpy as np
from prophet import Prophet
from sklearn.metrics import mean_squared_error

mlflow.set_tracking_uri("http://52.90.255.110:5000/")
mlflow.set_experiment("Demand_Forecasting_System")


results = []

for (store, item), group in df.groupby(['store', 'item']):

    group = group[['date', 'sales']]
    group = group.set_index('date').asfreq('D').fillna(0).reset_index()

    group = create_features(group)

    prophet_df = group.rename(columns={
        'date': 'ds',
        'sales': 'y'
    })

    with mlflow.start_run(run_name=f"store_{store}_item_{item}"):

        model = Prophet(
            yearly_seasonality=True,
            weekly_seasonality=True,
            daily_seasonality=False
        )

        model.add_regressor('lag_1')
        model.add_regressor('lag_7')
        model.add_regressor('rolling_7')

        model.fit(prophet_df)

        # Forecast
        future = prophet_df[['ds','lag_1','lag_7','rolling_7']]
        forecast = model.predict(future)

        rmse = np.sqrt(mean_squared_error(
            prophet_df['y'], forecast['yhat']
        ))

        # MLflow logging
        mlflow.log_param("store", store)
        mlflow.log_param("item", item)
        mlflow.log_param("model", "Prophet")
        mlflow.log_metric("rmse", rmse)

        model_name = f"demand_forecast_store_{store}_item_{item}"

        mlflow.prophet.log_model(
            model,
            artifact_path="model",
            registered_model_name=model_name
        )

        results.append((store, item, rmse))

        print(f"Store {store}, Item {item} ‚Üí RMSE: {rmse:.2f}")
```
