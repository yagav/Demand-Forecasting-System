```python
import pandas as pd

# -------- CONFIG --------
INPUT_CSV = "data\sales_2013.csv"          # original file
OUTPUT_CSV = "new_data\sales_2013.csv"    # filtered file
# ------------------------

# Load CSV
df = pd.read_csv(INPUT_CSV, parse_dates=["date"])

# Select first 4 stores (sorted for determinism)
first_4_stores = sorted(df["store"].unique())[:3]

# Filter data
filtered_df = df[df["store"].isin(first_4_stores)]

# Save filtered CSV
filtered_df.to_csv(OUTPUT_CSV, index=False)

print("Filtered stores:", first_4_stores)
print("Saved filtered file as:", OUTPUT_CSV)
```
